## Differenciation

Most methods for solving differential equations are *time-stepping* methods. The pairs $(t_i, y_i)$ are generated by taking small $t$ steps and approximating the value of the function $y$. This is perhaps best illustrated by Euler's method, which is the most basic time-stepping method. Fixing a small step size $h \gt 0$, we form the approximation at the $i$th step using the formula

$$
y_i = y_{i-1} + h\,f(t_{i-1}, y_{i-1})
$$

starting from the known initial value $y_0$.

```python
import numpy as np

def f(t, y):
    return -0.2*y

def euler(func, t_range, y0, step):
    t = [t_range[0]]
    y = [y0]
    i = 0
    while t[i] < t_range[1]:
        i += 1
        t.append(t[i-1] + step)
        y.append(y[i-1] + step*func(t[i-1], y[i-1]))
    return t, y

t_range = (0, 5)
t, y = euler(f, t_range, 50., .1)

assert type(t) == list
assert type(y) == list

def true_solution(t):
    return 50.*np.exp(-0.2*t)

array_t = np.array(t)
array_y = np.array(y)
array_T = true_solution(array_t)

assert type(array_t) == np.ndarray
assert type(array_y) == np.ndarray
assert type(true_T) == np.ndarray

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1,2, tight_layout=True)
ax1.plot(t, y)
ax1.set_xlabel(r"$t$")
ax1.set_ylabel(r"$T$")
ax1.set_title("Euler")
ax2.plot(array_t, np.abs(array_y - array_T))
ax2.set_xlabel(r"$t$")
ax2.set_ylabel("error")
ax2.set_title("error");
```
